```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
import pickle
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import MinMaxScaler
```

```python
data = pd.read_csv("players_22-1.csv")
```

```python
data.head()
```

```python
cols_required = ["value_eur", "age",	"potential",	"wage_eur",	"movement_reactions",	"defending",	"mentality_composure",	"skill_ball_control", "overall"]
```

```python
df = data[cols_required]
```

```python
df.head()
```

```python
df.info()
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
y = df['overall']

X = df.drop(columns='overall')
```

```python
# Load the trained Random Forest model
with open('random_forest_model.pkl', 'rb') as model_file:
    loaded_model = pickle.load(model_file)

# Load the fitted MinMaxScaler
with open('minmax_scaler.pkl', 'rb') as scaler_file:
    loaded_scaler = pickle.load(scaler_file)

# Scale the new data using the loaded MinMaxScaler
X_norm = loaded_scaler.transform(X)

# Use the loaded model to make predictions
predictions = loaded_model.predict(X_norm)

# Add predictions to the new data DataFrame
df['Predicted'] = predictions
```
